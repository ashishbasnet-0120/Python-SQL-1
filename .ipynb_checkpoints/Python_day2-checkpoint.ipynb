## conditional statement in pythhon

```python
# creating  a login functio using the if-else statement 

email=input("Enter your email, please:")
password=input("Enter your password:")

if email=="ashishbasnet0120@gmail.com" and password=="Ashish922#":
    print("sucessfully login")
    
elif email=="ashishbasnet0120@gmail.com" and  password!="Ashish922#":
    print("please check your password" )

elif email!="ashishbasnet0120@gmail.com" and  password=="Ashish922#":
    print("please check your email_username " )

else:
    print("wrong credentials ")
```

```python
# finding minimum of three numbers 

num1=float(input("enter the first num:"))
num2=float(input("enter the second num:"))
num3=float(input("enter the third num:"))

if num1<num2 and num1<num3:
    print(num1)

elif num2<num1 and num2<num3:
    print(num2)


else:
    print(num3)
```

```python
# building simple Atm machine using python


atm=int(input("""
Hi, this is atm machine,
1.Enter 1 for pin change
2.Enter 2 for balance check
3.Enter 3 for withdrawal of money
4.Enter 4 for exit 
"""))

if atm==1:
    pin=int(input("change your pin:"))
    print("successfully changed your pin, your pin is,",pin)

elif atm==2:
    print("your balance is NPR. 50,000")

elif atm==3:
    money=int(input("please enter your desirable amount of money to withdrawal:"))
    if money>=50000:
        print("sorry insufficient balance")
    else:
        print("suceessfully withdrawl,",money)

elif atm==4:
    exit=int(input("If you want to exit press 0:"))
    if exit==0:
        print("sucessfully exited")

    else:
        print("you are not able to exit the ATM system")

else:
    print(" you press the wrong key")
```
