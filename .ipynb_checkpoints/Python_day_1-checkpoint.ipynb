## Simple calculator using input function

```python
# calculator that performs add, subtract and multiplication 

num1=input("Enter the first number:")
num2=input("Enter the second number:")

add= num1+num2
diff= num1-num2
mul= num1*num2

print(add)
print(diff)
print(mul)

# showing problem because input function convert the data into string so usually string and string concatenate 
# but not subtract , multiply and so on so we have to do type conversion 
```

```python
# calculator that performs add, subtract and multiplication 

num1=float(input("Enter the first number:"))
num2=float(input("Enter the second number:"))

add= num1+num2
diff= num1-num2
mul= num1*num2

print(add)
print(diff)
print(mul)

```
