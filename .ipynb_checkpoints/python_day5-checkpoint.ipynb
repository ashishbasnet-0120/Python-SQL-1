```python
#creating a tuple 


t2=(2,3,"Ashish",(3,2),True,[30,40,60])


print(t2)
```

```python



# accesssing through indexing

t2=(2,3,"Ashish",(3,2),True,[30,40,60])

print(t2[5][2])

 
```

```python
# acessing through slicing
t2=(2,3,"Ashish",(3,2),True,[30,40,60])

print(t2[::2])
```



### editing / deletion is not possible because tuple is immutable but ca perform changing it in list

```python


# type conversion

t2=(2,3,"Ashish",(3,2),True,[30,40,60])
new_1=list(t2)
print(new_1)
```

```python



# tuple operation 

#arithmetic

t2=(2,3,4,6)
t3=(20,30,4)
print(t2+t3)
```

```python
#membership

t2=(2,3,4,6)
print( 2 not in t2)
```

```python

#loop
t2=(2,3,4,6)
for i in t2:
    print(i)
```

```python
#tuples methods 


t2=(2,3,4,6)

print(len(t2))
print(min(t2))
print(max(t2))
print(sorted(t2, reverse=True))

print(sum(t2))

```

```python
# methods

t2=(2,3,4,6)
print(t2.index(4))

print(t2.count(4))
```

```python




# tuple unpacking

a,b,c=(4,5,6)
print(a,b,c, sep="\n")
```

```python
# can be done for list 


a,b,c,d=[30,40,60,80,]
print(a,b,c,d)

```

```python
a,b,c,*not_intrested_to_unpack=[30,40,60,80,"Ashish",True]
print(a,b,c,not_intrested_to_unpack)
```
