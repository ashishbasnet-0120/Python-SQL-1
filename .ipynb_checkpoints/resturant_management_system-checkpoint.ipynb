```python
# resturant management system


menu={
    "momo":150,
    "coke":75,
    "water":20,
    "chowmein":120,
    "pizza":500,
    "sekuwa":350,
    "burger":700
    
}

print("""
Welcome To Local Thakali Resturant


Items we served are:
momo:150,
coke:75
water:20,
chowmein:120,
pizza:500,
sekuwa:350,
burger:700
   
    
               

""")
# .lower() converts the the input into lowercase


# dictoranary keys are strings
total=0
while True:
    order=input("enter the order by looking the menu:").lower()
    if order in menu:
        price=menu[order]
        total+=price
        print(f"your total price is {total}")

    else:
        print("item is unavailable")

    

    choice=input("Do you want to add the item:").lower()
    if choice=="no":
        break

print("thank you for visiting our resturant")
    


    

```
