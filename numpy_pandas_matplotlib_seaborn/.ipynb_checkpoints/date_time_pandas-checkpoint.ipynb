```python



# date range function 


# generate daily dates in a given range 

pd.date_range(start='2026-04-05',end='2026-05-01',freq='D')
```

```python


#D = Daily frequency
#So pandas generates one date per day
#It moves forward in steps of 1 day
```

```python


# alternate days in given range 
pd.date_range(start='2026-04-05',end='2026-05-01',freq='2D')
```

```python


# business days 


# remove saturday and sunday 

pd.date_range(start='2026-04-05',end='2026-05-01',freq='B')
```

```python



# one  day per week 

pd.date_range(start='2026-04-05',end='2026-05-01',freq='W-Thu')
```

```python


# hourly data 
pd.date_range(start='2026-04-05',end='2026-05-01',freq='H')
```

```python


pd.date_range(start='2026-04-05',end='2026-05-01',freq='6H')
```

```python


#month end

pd.date_range(start='2026-04-05',end='2026-05-01',freq='ME')
```

```python

#month start
pd.date_range(start='2026-04-05',end='2026-05-01',freq='MS')
```

```python


# using periods 

pd.date_range(start='2026-04-05',periods=30,freq='D')
```

```python



# to_datetime function 




x=pd.Series(['2023-01-03','2023-02-03','2023-04-03'])
x
```

```python


pd.to_datetime(x)
```

```python
pd.to_datetime(x).dt.year
```

```python
pd.to_datetime(x).dt.month
```

```python
# handling using error   

y=pd.Series(['2023-01-03','2023-02-03','2023-0400-03'])
y
```

```python
pd.to_datetime(y)
```

```python


# coerce means ignore skip

pd.to_datetime(y,errors='coerce')
```

```python
pd.to_datetime(y,errors='coerce').dt.day
```
