```python





# size

res=sns.load_dataset('tips')
res
```

```python


plt.scatter(res['tip'],res['total_bill'],marker='.',color='red')
plt.title('tip vs total_bill')
plt.xlabel('tip')
plt.ylabel('total_bill')
plt.show()
```

```python




# s represent size of each dot
# in case of the below dataset it represents with how many number of people came along with that customer 

plt.scatter(res['tip'],res['total_bill'],marker='.',color='red',s=res['size']*20)
plt.title('tip vs total_bill')
plt.xlabel('tip')
plt.ylabel('total_bill')
plt.show()
```

```python


# scatter plot using plt plot

# advantage:faster

# disadvantage: cant manipulate size and color
# o is for bi o plot
plt.plot(res['tip'],res['total_bill'],'o')
```

```python


# solution using xticks 

colors=['blue blue blue','light white','Dark greeen','pink pink pink pink pinkpink']
price=[10,20,40,70]



plt.bar(colors,price)
plt.title('colors price')
plt.xlabel('colors')
plt.ylabel('price')
plt.xticks(rotation=70)
plt.show()
```

```python


# assigning percentage
plt.pie(g['batsman_runs'],labels=g['batsman'],autopct='%1.0f%%')

plt.show()
```

```python
# basically 1.of means no decimal

# 1.1f means 1 decimal
```

```python

# explode means extracting the certain % outside  from the original position


plt.pie(g['batsman_runs'],labels=g['batsman'],autopct='%1.0f%%',explode=[0.1,0,0,0,0.1,0])

plt.show()
```

```python


# shadow  gives the 3d looks
plt.pie(g['batsman_runs'],labels=g['batsman'],autopct='%1.0f%%',explode=[0.1,0,0,0,0.1,0],shadow=True)

plt.show()
```
