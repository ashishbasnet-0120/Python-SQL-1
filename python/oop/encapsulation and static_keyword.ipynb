```python

"""
p1=Cordinate(0,0)
"""




class Cordinate:

    def __init__(self,x,y):
        self.x_cordinate=x
        self.y_cordinate=y
        

    def __str__(self):
        return '<{},{}>'.format(self.x_cordinate,self.y_cordinate)

    
    def distance(self,other): # here self is a first obj with x1 and y1 and other is a another obj with x2 and y2
        return ((self.x_cordinate-other.x_cordinate)**2+(self.y_cordinate-other.y_cordinate)**2)*0.5

    def distance_from_origin(self):
        return ((self.x_cordinate)**2+(self.y_cordinate)**2)*0.5
```

```python
p1=Cordinate(7,9)
p2=Cordinate(5,7)
print(p1)

print(p1.distance(p2))
p2.distance_from_origin()
```

```python




class Line:
    def __init__(self,A,B,C):
        self.A=A
        self.B=B
        self.C=C


    def __str__(self):
        return '{}X+{}Y+{}=0'.format(self.A,self.B,self.C)

    def check(line,point):
        if (line.A*point.X+line.B*point.Y+line.C)==0:
            return "it lies on a line"
        else:
            return "it doesn't lie on line "






class Point:
    def __init__(self,x,y):
        self.X=x
        self.Y=y
```

```python
p1=Line(3,4,5)
p2=Point(3,4)
print(p1)
```

```python
p1.check(p2)
```
