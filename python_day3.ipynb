```python
# pattern 
row=int(input("enter the amount of row:"))
for i in range(1,row+1):
    for j in range(1,i+1):
        print("*",end="")
    print("")
```

```python
# inverted half pyramid

half=int(input("enter the no of rows:"))

for i in range(half+1,0,-1):
    for j in range(0,i):
        print("*",end="")
    print("")
```

```python
for i in range(1,20):
    if i==10:
        break
    print(i)
```

```python
# prime number using break statement
start=int(input("enter the initial condition:"))
end=int(input("enter the stopping condition:"))

for i in range(start,end+1):
    if i<2:
        continue
    for j in range(2,i):
        if i%j==0:
            break
    else:
            print(i)

```
