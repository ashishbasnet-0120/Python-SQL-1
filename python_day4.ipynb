### list datatype in python

```python
# list can contain any type of data and nested list is possible
marks=[30,80.5,"good",True,[30,20,80]]
print(marks)
```

```python


#memory address of list

marks=[30,80.5,"good",True,[30,20,80]]
id(marks)


```

```python


#memory address of individual

marks=[30,80.5,"good",True,[30,20,80]]

id(marks[1])
```

```python
# list cab be duplicate 

sum=[3,2,1,10,1,1]
print(sum)

# here duplication of 1 is possible so list allows duplicate 
```

```python


#creation of list

#creation of empty list
ram=[]


# creation of 1-dimensional list

shyam=[3,4,5,6,7]


#creation of 2-dimensional list

hari=[1,2,3,["sita",True]]


#creation of 3-dimensional list
ramu=[[[1,3,6],["hari",30]]]


print(hari)
print(ramu)
```

```python




# acessing of list 

hari=[1,2,3,["sita",True]]
print(hari[0])


print(hari[3][1])




```

```python



# list methods 

#append
# add to the last of the list

a=[1,2,3,4,5]
a.append(["ram","sita"])
print(a)
```

```python



# extend
# same like append but it adds the items at last
#It does not create a nested list.

b=[1,2,3,4,5]
b.extend("sita")

print(b)
```

```python



#editable with indexing 

x=[300,400,False]
x[2]=True

print(x)
```

```python


#editable with slicing
x=[300,400,False]
x[1:3]="shyam",500
print(x)
```

```python


# inserting
x=[300,400,False]
x.insert(0,"pass")
print(x)

```

```python




#deletion

marks=[30,80.5,"good",True,[30,20,80]]
del marks[4]

print(marks)
```

```python



#pop method 

marks=[30,80.5,"good",True,[30,20,80]]
print(marks.pop())
print(marks)
```

```python


#clear

marks=[30,80.5,"good",True,[30,20,80]]
marks.clear()
print(marks)
```

```python
# len/min/max/sorted 

marks=[30,80.5,"good",True,[30,20,80]]

print(len(marks))


#print(min(marks))   error comes because python doesn't know how to compare string and int



com=[4,5,10.5,16]
print(min(com))

print(max(com))


print(sorted(com,reverse=True))

print(sorted(com))
```

```python



## operations on list


# arithmetic operation +,*

L1=[1,2,3,4,5]
L2=[6,7,8,9]

print(L1+L2)

L3=L1+L2
print(L3)

#multiplication

L1*3


# we cant multiply two list it is only possible for single list



```

```python
#membership operation


marks=[30,80.5,"good",True,[30,20,80]]

print(30 in marks)

print(False not in marks)
```

```python



# loop in list

marks=[30,80.5,"good",True,[30,20,80]]

for z in marks:
    print(z)
```

```python
# list comprehension 

#adding 1 to 10 in a list
l=[]
for i in range(1,11):
    l.append(i)
print(l)



```

```python
# through list comprehension
new_list=[i for i in range(1,11)]
print(new_list)

```

```python
#multiplication

new=[3,2,1]
mul=-2
net=[]
for i in new:
    result=i*mul
    net.append(result)

print(net)
```

```python
# squaring a number 
num=int(input("enter the number you want to square:"))
num1=int(input("enter the another you want to square:"))

square=[num,num1]
result=[]
for i in square:
    result.append(i**2)

print(result)
```

```python
# through list comprehension

num=int(input("enter the number you want to square:"))
num1=int(input("enter the another you want to square:"))
square=[num,num1]
new_list=[i**2 for i in square ]
print(new_list)
```

```python




#using condition
hello=[0,1,2,10,15,16,20,13,18]

new_list=[i for i in hello if i%2==0]
print(new_list)
```

```python



#zip function in python

l1=[1,2,3,"Ashish"]
l2=[20,40,60,40]
zip(l1,l2)

list(zip(l1,l2))
```

```python
# if they have different length

l1=[1,2,3,"Ashish","manish",80]
l2=[20,40,60,40]
zip(l1,l2)

tuple(zip(l1,l2))
dict(zip(l1,l2))
```
