```python






#sets in python

#creating a empty set

s1=set()
print(type(s1))

```

```python
# set is ordered

s2={4,5,6}
s4={6,5,4}

print(s2==s4)
```

```python



#1 dimensional set
# it does not allow duplicates 
s1={1,2,3,4,1}

print(s1)
```

```python




#2 dimensional set

# list is not allowed in a set because list is mutable (unhashable)
# dict is not allowed because dict is mutable (unhashable)
# set is not allowed because set is mutable (unhashable)
# Only immutable (hashable) objects can be elements of a set

s5={1,2,3,3,40.2,("Ashish", 30)}
print(s5)
```

```python



# Accessing item in a set

# it is not possible to access because it is unorderd so, it does not have any index so, you cant perform indexing and slicing
```

```python
# you can't assign the values in sets because it does not have any index
```

```python


#addding values 

s1={1,2,3,4,1}
s1.add(10) # it will add 1 item
print(s1)



s1.update(["Ashish",60])# adding multiple items but we should pass in a list form
print(s1)
```

```python



#deleting from set
s7={1, 2, 3, 4, 10, 'Ashish', 60}
s7.discard(60)
print(s7)
```

```python


s7={1, 2, 3, 4, 10, 'Ashish', 60}
s7.remove(60)
print(s7)
```

```python


s7={1, 2, 3, 4, 10, 'Ashish', 60}
s7.pop()# randomly remove 1 item
print(s7)
```

```python
s7={1, 2, 3, 4, 10, 'Ashish', 60}
s7.clear()# clears the set
print(s7)
```

```python



# set operations 

# union
s1={1, 2, 3, 4, 10, 'Ashish', 60}
s2={1,2,3,4,1}

print(s1|s2)
```

```python

# set difference
s1={1, 2, 3, 4, 10, 'Ashish', 60}
s2={1,2,3,4,1}

print(s1-s2)

```

```python

# intersection
s1={1, 2, 3, 4, 10, 'Ashish', 60}
s2={1,2,3,4,1}

print(s1&s2)
```

```python


#symmetric differemce ^
s1={1, 2, 3, 4, 10, 'Ashish', 60}
s2={1,2,3,4,1}

print(s1^s2)
```

```python

# menbership
s1={1, 2, 3, 4, 10, 'Ashish', 60}

print("Ashish" not in s1)
```

```python


# methods in sets 

s1={1, 2, 3, 4, 10,  60}
print(len(s1))
print(min(s1))
print(max(s1))
print(sum(s1))


print(sorted(s1, reverse=True))
```
