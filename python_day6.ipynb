```python
# higher order function 
# if a function takes a function as a paramater then its a HOF

# example 

num1=int(input("enter the first num:"))
num2=int(input("enter the second num:"))

operands=[num1,num2]
def add(a,b):
    return a+b

def mul(a,b):
    return a*b


def trans(function,oper):
    transformed= function(oper[0],oper[1])
   

    print(transformed)

trans(add,operands)
```

```python
# we can do other function without def other function with the help of lambda function

# higher order function 
# if a function takes a function as a paramater then its a HOF

# example 

num1=int(input("enter the first num:"))
num2=int(input("enter the second num:"))

operands=[num1,num2]
def add(a,b):
    return a+b

def mul(a,b):
    return a*b


def trans(function,oper):
    transformed= function(oper[0],oper[1])
   

    print(transformed)

trans(lambda x,y:x*y,operands)
```
